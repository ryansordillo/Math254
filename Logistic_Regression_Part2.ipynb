Q8: Use the `glm()` command to fit a logistic model to predict the acceptance probability usinig `GPA` and `Sex` as predictors

```r
# prompt: Use the glm() command to fit a logistic model to predict the acceptance probability usinig GPA and Sex as predictors

mymodel <- glm(Accept_code ~ GPA + Sex + GPA*Sex, data = mydata, family = "binomial")
summary(mymodel)
```

Q9: Write down equations of the two fitted logistic curves.

```r
# prompt: Q9: Write down equations of the two fitted logistic curves.

# Extract coefficients for the logistic regression model
coefficients <- coef(mymodel)

# Equation for Females:
# logit(P(Acceptance|Female)) = Intercept + GPA_Coefficient * GPA + Female_Coefficient + Interaction_Coefficient * GPA
# P(Acceptance|Female) = exp(Intercept + GPA_Coefficient * GPA + Female_Coefficient + Interaction_Coefficient * GPA) / (1 + exp(Intercept + GPA_Coefficient * GPA + Female_Coefficient + Interaction_Coefficient * GPA))

# Equation for Males:
# logit(P(Acceptance|Male)) = Intercept + GPA_Coefficient * GPA
# P(Acceptance|Male) = exp(Intercept + GPA_Coefficient * GPA) / (1 + exp(Intercept + GPA_Coefficient * GPA))

# Replace placeholders with actual coefficients from the model summary
# Example (replace with your model's coefficients):

intercept <- coefficients[1]
gpa_coef <- coefficients[2]
female_coef <- coefficients[3]
interaction_coef <- coefficients[4]

cat("Equation for Females:\nP(Acceptance|Female) = exp(", intercept, " + ", gpa_coef, "*GPA + ", female_coef, " + ", interaction_coef,"*GPA) / (1 + exp(", intercept, " + ",gpa_coef,"*GPA + ", female_coef, " + ", interaction_coef,"*GPA))\n\n")

cat("Equation for Males:\nP(Acceptance|Male) = exp(", intercept, " + ", gpa_coef, "*GPA) / (1 + exp(", intercept, " + ",gpa_coef,"*GPA))\n")
```
